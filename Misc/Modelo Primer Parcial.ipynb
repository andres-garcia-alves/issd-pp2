```python
# ============================================================
# 🌌 MISIÓN GALÁCTICA: INGENIERÍA DE DATOS NASA
# Autor de la presentación: Julio Paredes
# Proyecto: Gemelo Digital para Mantenimiento Predictivo
# Dataset: NASA CMAPSS - Motores Turbofán
# ============================================================

# =========================
# 0) IMPORTACIONES GENERALES
# =========================
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import kagglehub
import ipywidgets as widgets

from IPython.display import display, HTML, clear_output
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn import tree
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import IsolationForest

# ============================================================
# 1) PORTADA / PRESENTACIÓN INICIAL EN HTML
# ============================================================
portada_html = """
<style>
    @import url('https://fonts.googleapis.com/css2?family=Orbitron:wght@500;700;800&family=Rajdhani:wght@400;500;600;700&display=swap');

    .galaxy-box {
        max-width: 1200px;
        margin: 20px auto;
        padding: 35px;
        border-radius: 22px;
        background:
            radial-gradient(circle at top left, rgba(0, 242, 255, 0.12), transparent 25%),
            radial-gradient(circle at top right, rgba(255,255,255,0.08), transparent 18%),
            linear-gradient(135deg, #030814 0%, #07182d 55%, #0a2240 100%);
        color: #eef8ff;
        font-family: 'Rajdhani', sans-serif;
        border: 1px solid rgba(0,242,255,0.25);
        box-shadow: 0 0 30px rgba(0,180,255,0.12);
    }
    .galaxy-title {
        font-family: 'Orbitron', sans-serif;
        font-size: 38px;
        color: #ffffff;
        margin-bottom: 10px;
        text-align: center;
    }
    .galaxy-subtitle {
        text-align: center;
        font-size: 22px;
        color: #8be9ff;
        margin-bottom: 20px;
        font-weight: 700;
    }
    .galaxy-author {
        text-align: center;
        font-size: 22px;
        color: #ffd166;
        font-weight: 700;
        margin-bottom: 20px;
    }
    .galaxy-text {
        font-size: 20px;
        line-height: 1.65;
        color: #eaf6ff;
    }
    .galaxy-group {
        margin-top: 20px;
        background: rgba(255,255,255,0.04);
        border-left: 4px solid #00f2ff;
        padding: 16px;
        border-radius: 10px;
        font-size: 18px;
    }
</style>

<div class="galaxy-box">
    <div class="galaxy-title">🌌 MISIÓN GALÁCTICA: INGENIERÍA DE DATOS NASA</div>
    <div class="galaxy-subtitle">Gemelo Digital para Mantenimiento Predictivo en Motores Turbofán</div>
    <div class="galaxy-author">✍️ Autor de la presentación: Julio Paredes</div>

    <div class="galaxy-text">
        Este proyecto desarrolla una propuesta de <b>mantenimiento predictivo</b> inspirada en la lógica de un
        <b>gemelo digital</b>, utilizando el dataset <b>CMAPSS de la NASA</b>.
        A partir de datos de sensores, condiciones operativas y ciclos de vida de motores turbofán,
        se construye un flujo de trabajo que combina <b>auditoría de datos</b>,
        <b>ingeniería de variables</b>, <b>modelado descriptivo</b> y
        <b>detección proactiva de anomalías</b>.
    </div>

    <div class="galaxy-group">
        <b>Grupo 12:</b> Ciuró Casas, Marcel — Guisasola Uriarte, Jon — Paredes Rojas, Julio Alberto —
        Trilles Barreda, Vicent.<br><br>
        <i>Todos los integrantes de este grupo han participado de manera equitativa, activa y colaborativa
        en el desarrollo integral de este proyecto.</i>
    </div>
</div>
"""
display(HTML(portada_html))

# ============================================================
# 2) DESCARGA DEL DATASET NASA CMAPSS DESDE KAGGLEHUB
# ============================================================

# Descarga la última versión disponible del dataset desde KaggleHub
dataset_path = kagglehub.dataset_download("bishals098/nasa-turbofan-engine-degradation-simulation")

print("📂 Ruta general descargada por KaggleHub:")
print(dataset_path)

# Ruta específica donde suelen quedar los archivos del dataset
# Si KaggleHub cambia la carpeta exacta, podés adaptar esta línea
path = "/root/.cache/kagglehub/datasets/bishals098/nasa-turbofan-engine-degradation-simulation/versions/1"

# Carga del conjunto de entrenamiento
train = pd.read_csv(os.path.join(path, "train_FD001.txt"), sep=r"\s+", header=None)

# Carga del conjunto de prueba
test = pd.read_csv(os.path.join(path, "test_FD001.txt"), sep=r"\s+", header=None)

# Carga del archivo RUL del conjunto de test
rul = pd.read_csv(os.path.join(path, "RUL_FD001.txt"), sep=r"\s+", header=None)

print("\n✅ Archivos cargados correctamente.")
print("Shape train:", train.shape)
print("Shape test :", test.shape)
print("Shape rul  :", rul.shape)

# Vista preliminar del dataset
display(train.head())

# ============================================================
# 3) DICCIONARIO DE DATOS / EXPLICACIÓN DEL DATASET
# ============================================================
diccionario_html = """
<style>
    .cmapss-box {
        max-width: 1200px;
        margin: 20px auto;
        background: linear-gradient(135deg, #08111f 0%, #102846 100%);
        color: #f3faff;
        padding: 30px;
        border-radius: 18px;
        font-family: Arial, sans-serif;
        border: 1px solid rgba(0,242,255,0.2);
        box-shadow: 0 0 18px rgba(0,180,255,0.08);
    }
    .cmapss-box h1, .cmapss-box h2 {
        color: #7fe7ff;
    }
    .cmapss-box p, .cmapss-box li {
        font-size: 18px;
        line-height: 1.6;
    }
    .cmapss-box table {
        width: 100%;
        border-collapse: collapse;
        margin-top: 18px;
        background: rgba(255,255,255,0.03);
    }
    .cmapss-box th, .cmapss-box td {
        border: 1px solid rgba(255,255,255,0.12);
        padding: 10px;
        text-align: left;
    }
    .cmapss-box th {
        background: rgba(0,242,255,0.10);
        color: #ffffff;
    }
</style>

<div class="cmapss-box">
    <h1>🚀 CMAPSS – Motor Turbofán</h1>
    <h2>NASA · Predictive Maintenance</h2>

    <p>
        Este panel resume el <b>diccionario de datos</b> del conjunto <b>C-MAPSS</b>.
        Cada fila representa un <b>ciclo de operación</b> de un motor turbofán simulado por la NASA.
        Los sensores y ajustes operativos permiten estimar la <b>vida útil remanente (RUL)</b> de cada motor.
    </p>

    <h2>🧭 Identificación y tiempo</h2>
    <p><b>unit</b>, <b>cycle</b> (Clave temporal)</p>
    <ul>
        <li><b>unit</b>: identifica el motor (unidad 1, 2, …, N).</li>
        <li><b>cycle</b>: número de ciclo u “hora de vuelo simulada”.</li>
        <li>Cada motor empieza en 1 y aumenta hasta la falla.</li>
        <li>El <b>RUL</b> se calcula a partir del último ciclo de cada unidad.</li>
    </ul>

    <h2>⚙️ Condición operativa</h2>
    <p><b>op_setting_1</b> · <b>op_setting_2</b> · <b>op_setting_3</b></p>
    <ul>
        <li>Representan el entorno y la carga de operación del motor.</li>
        <li>Influyen en la velocidad de degradación del sistema.</li>
        <li>Condiciones más exigentes → mayor desgaste operativo.</li>
    </ul>

    <h2>📡 Sensores</h2>
    <p><b>sensor_1 … sensor_21</b> (21 canales)</p>
    <ul>
        <li>Cada sensor registra magnitudes físicas del motor.</li>
        <li>Incluyen señales asociadas con presión, temperatura, flujo y comportamiento mecánico.</li>
        <li>Sus cambios de tendencia ayudan a anticipar fallas.</li>
    </ul>

    <h2>🎯 Propósito del dataset</h2>
    <p><b>Remaining Useful Life (RUL)</b></p>
    <ul>
        <li>El objetivo es estimar cuántos ciclos le quedan al motor antes de fallar.</li>
        <li>Esto permite programar mantenimiento ni demasiado tarde ni demasiado temprano.</li>
    </ul>

    <h2>📊 Tabla resumida de variables</h2>
    <table>
        <tr>
            <th>N°</th>
            <th>Nombre</th>
            <th>Tipo</th>
            <th>Descripción breve</th>
        </tr>
        <tr>
            <td>0</td>
            <td>unit</td>
            <td>Entero</td>
            <td>ID del motor turbofán</td>
        </tr>
        <tr>
            <td>1</td>
            <td>cycle</td>
            <td>Entero</td>
            <td>Ciclo de operación (tiempo)</td>
        </tr>
        <tr>
            <td>2</td>
            <td>op_setting_1</td>
            <td>Real</td>
            <td>Ajuste o condición operativa 1</td>
        </tr>
        <tr>
            <td>3</td>
            <td>op_setting_2</td>
            <td>Real</td>
            <td>Ajuste o condición operativa 2</td>
        </tr>
        <tr>
            <td>4</td>
            <td>op_setting_3</td>
            <td>Real</td>
            <td>Ajuste o condición operativa 3</td>
        </tr>
        <tr>
            <td>5–25</td>
            <td>sensor_1 … sensor_21</td>
            <td>Real</td>
            <td>Lecturas de 21 sensores físicos del motor</td>
        </tr>
    </table>
</div>
"""
display(HTML(diccionario_html))

# ============================================================
# 4) FASE 1: EVALUACIÓN DE LA CALIDAD DE LOS DATOS
# ============================================================

# Se asignan nombres interpretables a las columnas según la estructura conocida del dataset
nombres_columnas = (
    ['engine_id', 'cycle', 'setting_1', 'setting_2', 'setting_3']
    + [f'sensor_{i}' for i in range(1, 22)]
)

train.columns = nombres_columnas
test.columns = nombres_columnas

# ---- 4.1 Identificación de nulos y duplicados ----
nulos_totales = train.isnull().sum().sum()
duplicados = train.duplicated().sum()

# ---- 4.2 Sensores constantes ----
# Un sensor constante no aporta variabilidad y puede eliminarse
sensores_constantes = [col for col in train.columns if train[col].nunique(dropna=False) <= 1]

# ---- 4.3 Detección simple de outliers ----
# Regla estadística: media ± 3*desviación estándar
num = train.select_dtypes(include=[np.number])
stats = num.describe().T

std = stats['std'].replace(0, np.nan)
lower = stats['mean'] - 3 * std
upper = stats['mean'] + 3 * std

outliers = ((num.lt(lower, axis=1)) | (num.gt(upper, axis=1))).sum().sum()

# ---- 4.4 Reporte final de la fase 1 ----
print("\n🛰️ FASE 1: DIAGNÓSTICO DE CALIDAD")
print("-" * 50)
print(f"🔍 Nulos totales:              {nulos_totales}")
print(f"👯 Filas duplicadas:           {duplicados}")
print(f"📉 Sensores constantes:        {len(sensores_constantes)}")
if len(sensores_constantes) > 0:
    print("   ->", ", ".join(sensores_constantes))
print(f"☄️ Outliers (media ± 3*std):   {outliers}")
print("-" * 50)
print("✅ Calidad validada. Datos listos para el preprocesamiento.")

# ============================================================
# 5) FASE 2: INGENIERÍA DEL GEMELO DIGITAL
# ============================================================

# ---- 5.1 Eliminación de sensores constantes ----
# Estos sensores fueron detectados como constantes o poco útiles para el modelado
sensores_constantes_nombres = [
    'setting_3', 'sensor_1', 'sensor_5', 'sensor_10',
    'sensor_16', 'sensor_18', 'sensor_19'
]

train_cleaned = train.drop(columns=sensores_constantes_nombres)

# ---- 5.2 Construcción del objetivo de minería: RUL ----
# Para cada motor, el último ciclo observado representa el fin de su vida útil en train
max_cycle = train_cleaned.groupby('engine_id')['cycle'].transform('max')
train_cleaned['RUL'] = max_cycle - train_cleaned['cycle']

# Variable binaria de riesgo:
# 1 = el motor está cerca de la falla (<= 30 ciclos restantes)
# 0 = operación normal
train_cleaned['label'] = (train_cleaned['RUL'] <= 30).astype(int)

# ---- 5.3 Selección de features relevantes ----
# Se renombran algunos sensores elegidos para trabajar con nombres más claros
renombrado_features = {
    'sensor_11': 'feature1',  # temperatura total salida HPC
    'sensor_2':  'feature2',  # temperatura total salida LPC
    'sensor_7':  'feature3'   # presión total salida HPC
}

train_cleaned = train_cleaned.rename(columns=renombrado_features)

# ---- 5.4 Normalización ----
# Escalado Min-Max para que las variables queden entre 0 y 1
scaler = MinMaxScaler()
cols_a_escalar = ['feature1', 'feature2', 'feature3']
train_cleaned[cols_a_escalar] = scaler.fit_transform(train_cleaned[cols_a_escalar])

# ---- 5.5 Reporte de la fase 2 ----
print("\n🛠️ FASE 2: CONFIGURACIÓN DEL GEMELO DIGITAL")
print("-" * 65)
print("POR QUÉ SE USAN ESTOS SENSORES COMO FEATURES:")
print("1. feature1: alta sensibilidad al desgaste del compresor de alta presión.")
print("2. feature2: captura degradación en etapas de baja presión.")
print("3. feature3: permite observar pérdida de capacidad de presurización.")
print("-" * 65)
print("✅ Variable 'label' creada correctamente.")
print("📊 Vista previa del dataset listo para modelado:")
display(train_cleaned[['engine_id', 'cycle', 'feature1', 'feature2', 'feature3', 'RUL', 'label']].head())

# ============================================================
# 6) FASE 3: MODELO DESCRIPTIVO CON ÁRBOL DE DECISIÓN
# ============================================================

# ---- 6.1 Preparación de variables ----
X = train_cleaned[['feature1', 'feature2', 'feature3']]
y = train_cleaned['label']

# ---- 6.2 Entrenamiento del árbol ----
# max_depth=3 para conservar interpretabilidad
modelo_arbol = DecisionTreeClassifier(max_depth=3, random_state=42)
modelo_arbol.fit(X, y)

# ---- 6.3 Extracción de reglas lógicas ----
reglas_texto = export_text(modelo_arbol, feature_names=['feature1', 'feature2', 'feature3'])

# ---- 6.4 Predicciones y métricas ----
y_pred = modelo_arbol.predict(X)
reporte_metricas = classification_report(y, y_pred, target_names=['Sano', 'Riesgo'])

print("\n🧠 FASE 3: ANÁLISIS DESCRIPTIVO DEL GEMELO DIGITAL")
print("-" * 65)
print("📜 REGLAS DE DECISIÓN GENERADAS:")
print(reglas_texto)
print("-" * 65)
print("📊 MÉTRICAS DE RENDIMIENTO:")
print(reporte_metricas)

# ---- 6.5 Visualización del árbol ----
plt.figure(figsize=(20, 10))
tree.plot_tree(
    modelo_arbol,
    feature_names=['feature1', 'feature2', 'feature3'],
    class_names=['Sano', 'Riesgo'],
    filled=True,
    rounded=True,
    fontsize=12
)
plt.title("MAPA DE DECISIÓN: PATRONES DE FALLA EN MOTORES NASA")
plt.show()

# ============================================================
# 7) MATRIZ DE CONFUSIÓN Y MÉTRICAS FINALES
# ============================================================

cm = confusion_matrix(y, y_pred)
labels = ['Sano (Operación)', 'Riesgo (Mantenimiento)']

fig_cm = px.imshow(
    cm,
    text_auto=True,
    aspect="auto",
    labels=dict(x="Predicción del Modelo", y="Estado Real", color="Cantidad"),
    x=labels,
    y=labels,
    color_continuous_scale='RdYlGn_r'
)

fig_cm.update_layout(
    title="<b>MATRIZ DE CONFUSIÓN: DIAGNÓSTICO DEL GEMELO DIGITAL</b>",
    template='plotly_dark',
    width=700,
    height=550
)
fig_cm.show()

print("\n📋 MÉTRICAS DE RENDIMIENTO DETALLADAS")
print("-" * 50)
print(classification_report(y, y_pred, target_names=labels))
print("-" * 50)
print("✅ Fase 3 completada: árbol entrenado, interpretado y evaluado.")

# ============================================================
# 8) ANÁLISIS TEXTUAL DE RESULTADOS
# ============================================================

print("""
📈 EVALUACIÓN DEL MODELO Y ANÁLISIS DE RESULTADOS
-------------------------------------------------
Los resultados obtenidos muestran que el Gemelo Digital logra distinguir entre
motores en condiciones normales y motores que ingresan en una zona de riesgo operativo.

Interpretación general:
- Los verdaderos positivos indican motores en riesgo correctamente detectados.
- Los verdaderos negativos muestran motores sanos clasificados correctamente.
- Las falsas alarmas representan inspecciones posiblemente innecesarias.
- Las fallas no detectadas marcan el margen de mejora del sistema.

En conjunto:
- Precision mide qué tan confiable es una alerta emitida por el sistema.
- Recall mide la capacidad real de detectar fallas.
- F1-Score resume el equilibrio entre ambas métricas.

Esto sugiere que el modelo puede funcionar como una base descriptiva útil para
mantenimiento predictivo y monitoreo industrial.
""")

# ============================================================
# 9) FASE 4: MODELO PROACTIVO - DETECCIÓN DE ANOMALÍAS
# ============================================================

# Se utiliza Isolation Forest para detectar observaciones atípicas
iso_forest = IsolationForest(contamination=0.05, random_state=42)

# Variables utilizadas para la detección
X_anom = train_cleaned[['feature1', 'feature2', 'feature3']]

# Predicción de anomalías:
#  1  -> normal
# -1  -> anomalía
train_cleaned['anomaly_score'] = iso_forest.fit_predict(X_anom)
train_cleaned['estado_proactivo'] = train_cleaned['anomaly_score'].map({1: 'Normal', -1: 'Anomalía'})

# Conteo total de anomalías detectadas
total_anomalias = (train_cleaned['anomaly_score'] == -1).sum()

print("\n🚀 FASE 4: DETECCIÓN PROACTIVA FINALIZADA")
print("-" * 55)
print(f"🔍 Registros analizados: {len(train_cleaned)}")
print(f"⚠️ Anomalías detectadas: {total_anomalias}")
print("-" * 55)

# ---- 9.1 Visualización 3D del espacio sensorial ----
fig_anom = px.scatter_3d(
    train_cleaned.head(5000),
    x='feature1',
    y='feature2',
    z='feature3',
    color='estado_proactivo',
    title="<b>DETECCIÓN DE ANOMALÍAS: ESPACIO SENSORIAL TRIDIMENSIONAL</b>",
    labels={
        'feature1': 'Presión / Temperatura crítica 1',
        'feature2': 'Temperatura crítica 2',
        'feature3': 'Presión crítica 3'
    },
    color_discrete_map={'Normal': '#00CC96', 'Anomalía': '#EF553B'},
    opacity=0.6
)

fig_anom.update_layout(template='plotly_dark', margin=dict(l=0, r=0, b=0, t=50))
fig_anom.show()

# ============================================================
# 10) CONSOLA DE MANDO INTERACTIVA DEL GEMELO DIGITAL
# ============================================================

# ---- 10.1 Estilo visual del panel ----
style_css = """
<style>
    .nasa-container {
        background-color: #000b1a;
        border: 2px solid #00f2ff;
        border-radius: 12px;
        padding: 20px;
        font-family: 'Share Tech Mono', monospace;
        color: #00f2ff;
        box-shadow: 0 0 15px #005f6b;
    }
    .nasa-header {
        text-align: center;
        color: #1ae891;
        text-transform: uppercase;
        letter-spacing: 2px;
        border-bottom: 1px solid #005f6b;
        padding-bottom: 10px;
    }
    .stat-card {
        background: rgba(0, 242, 255, 0.05);
        border: 1px solid #005f6b;
        border-radius: 8px;
        padding: 10px;
        text-align: center;
        width: 30%;
    }
</style>
"""

# ---- 10.2 Controles interactivos ----
slider_contaminacion = widgets.FloatSlider(
    value=0.05,
    min=0.01,
    max=0.20,
    step=0.01,
    description='🛰️ SENSIBILIDAD:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='95%')
)

btn_sincronizar = widgets.Button(
    description='🚀 SINCRONIZAR GEMELO DIGITAL',
    button_style='info',
    layout=widgets.Layout(width='100%', height='50px'),
    icon='refresh'
)

output_grafico = widgets.Output()
output_metricas = widgets.Output()

# ---- 10.3 Función principal del tablero ----
def sincronizar_sistema(b):
    with output_grafico:
        clear_output(wait=True)

        # Toma el umbral del slider y reentrena el modelo de anomalías
        umbral = slider_contaminacion.value
        modelo_iso = IsolationForest(contamination=umbral, random_state=42)

        X_vars = train_cleaned[['feature1', 'feature2', 'feature3']]
        predicciones = modelo_iso.fit_predict(X_vars)

        # Se arma una muestra para visualización fluida
        df_plot = train_cleaned.head(3000).copy()
        df_plot['Estado'] = ['⚠️ ANOMALÍA' if p == -1 else '✅ NORMAL' for p in predicciones[:3000]]

        # Gráfico 3D actualizado
        fig = px.scatter_3d(
            df_plot,
            x='feature1',
            y='feature2',
            z='feature3',
            color='Estado',
            title=f"ESTADO DEL GEMELO DIGITAL (Umbral: {umbral:.0%})",
            color_discrete_map={'✅ NORMAL': '#1ae891', '⚠️ ANOMALÍA': '#ff4500'},
            opacity=0.6
        )
        fig.update_layout(
            template='plotly_dark',
            paper_bgcolor='rgba(0,0,0,0)',
            margin=dict(l=0, r=0, b=0, t=40)
        )
        fig.show()

    with output_metricas:
        clear_output(wait=True)

        anomalias_totales = (predicciones == -1).sum()
        f1_base = 0.78  # referencia del modelo descriptivo

        if umbral < 0.06:
            interpretacion = "MODO CONSERVADOR: seguridad estándar."
            color_int = "#1ae891"
        elif umbral <= 0.12:
            interpretacion = "MODO EQUILIBRADO: monitoreo recomendado."
            color_int = "#3bc9ff"
        else:
            interpretacion = "MODO ALTA SENSIBILIDAD: alerta proactiva máxima."
            color_int = "#ff4500"

        display(HTML(f"""
        {style_css}
        <div style="display: flex; justify-content: space-around; margin-top: 15px;">
            <div class="stat-card">
                <p style="font-size:10px; margin:0;">ESTADO</p>
                <p style="color:#1ae891; font-weight:bold;">OPERATIVO</p>
            </div>
            <div class="stat-card">
                <p style="font-size:10px; margin:0;">ANOMALÍAS</p>
                <p style="color:#ff4500; font-weight:bold;">{anomalias_totales} U.</p>
            </div>
            <div class="stat-card">
                <p style="font-size:10px; margin:0;">F1-SCORE</p>
                <p style="color:#3bc9ff; font-weight:bold;">{f1_base}</p>
            </div>
        </div>
        <div style="margin-top:15px; padding:10px; border-left:4px solid {color_int}; background:rgba(255,255,255,0.05);">
            <p style="font-size:13px; color:{color_int}; margin:0;">
                <b>INTERPRETACIÓN TÉCNICA:</b> {interpretacion}
            </p>
        </div>
        """))

# ---- 10.4 Enlace del botón con la función ----
btn_sincronizar.on_click(sincronizar_sistema)

# ---- 10.5 Estructura final del panel ----
header_html = widgets.HTML(
    f"{style_css}<div class='nasa-container'><div class='nasa-header'><h2>📡 PANEL DE CONTROL GEMELO DIGITAL</h2></div>"
)
footer_html = widgets.HTML("</div>")

ui_controles = widgets.VBox([
    widgets.Label("Ajuste de Nivel de Contaminación (Umbral de Riesgo)",
                  layout=widgets.Layout(margin='10px 0 0 10px')),
    slider_contaminacion,
    btn_sincronizar
])

dashboard_final = widgets.VBox([
    header_html,
    ui_controles,
    output_metricas,
    output_grafico,
    footer_html
])

display(dashboard_final)

# Inicialización automática del panel
sincronizar_sistema(None)

# ============================================================
# 11) CIERRE INTERPRETATIVO
# ============================================================
print("""
🌌 CONCLUSIÓN GENERAL
-------------------------------------------------
Este flujo integra calidad de datos, construcción de RUL, explicabilidad mediante
árbol de decisión y monitoreo proactivo con Isolation Forest.

El resultado es un prototipo funcional de gemelo digital analítico que permite:
- comprender patrones de degradación,
- anticipar comportamientos anómalos,
- apoyar decisiones de mantenimiento predictivo,
- y comunicar resultados con una interfaz visual clara e interactiva.
""")
```
